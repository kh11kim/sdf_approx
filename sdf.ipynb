```python
import jax.numpy as jnp
import jax
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pandas as pd
```

```python
def sdf_circle(xy, center_xy=[0,0], radius=0.8):
    return jnp.sqrt((xy[0] - center_xy[0])**2 + (xy[1] - center_xy[1])**2) - radius
def sdf_square(xy):
  q = jnp.abs(xy) - jnp.array([0.75, 0.75])
  return jnp.linalg.norm(jnp.maximum(q, 0)) + jnp.minimum(jnp.maximum(q[0], jnp.maximum(q[1], q[2])), 0)

sdf_circle_batch = jax.vmap(sdf_circle)
sdf_square_batch = jax.vmap(sdf_square)
```

```python
# Set up grid size and circle parameters
n = 50
x = jnp.linspace(-1.5,1.5,n)
y = jnp.linspace(-1.5,1.5,n)
X, Y = jnp.meshgrid(x,y)
xy = jnp.hstack([X.reshape(-1,1), Y.reshape(-1,1)])
```

```python
sdf_cm = colors.LinearSegmentedColormap.from_list('SDF', [(0,'#eff3ff'),(0.5,'#3182bd'),(0.5,'#31a354'),(1,'#e5f5e0')], N=256)

# Calculate distance function and plot
Z_circle = sdf_circle_batch(xy).reshape(n, -1)
Z_square = sdf_square_batch(xy).reshape(n, -1)
levels = jnp.linspace(-1.5, 1.5, 21)
fig, axes = plt.subplots(1, 2, figsize=[8, 3.5])
axes[0].contourf(X, Y, Z_circle, levels=levels, cmap=sdf_cm)
axes[1].contourf(X, Y, Z_square, levels=levels, cmap=sdf_cm)

plt.show()
```

```python
df_circle = pd.DataFrame()
xy = np.random.uniform([-1.5, -1.5], [1.5, 1.5], size=(10000, 2))
df_circle.loc[:,["x", "y"]] = xy
df_circle.loc[:, "d"] = sdf_circle_batch(xy)
```

```python
df_square = pd.DataFrame()
xy = np.random.uniform([-1.5, -1.5], [1.5, 1.5], size=(10000, 2))
df_square.loc[:,["x", "y"]] = xy
df_square.loc[:, "d"] = sdf_square_batch(xy)
```

```python
df_circle.to_csv("circle.csv")
df_square.to_csv("square.csv")
```

```python
plt.scatter(*xy.T, c=df_square["d"], cmap=sdf_cm)
```
